# EDA
Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets for patterns, and anomalies (outliers), and form hypotheses based on our understanding of the dataset and summarize their main characteristics, often employing data visualization methods. It is an important step in any Data Analysis or Data Science project. It helps determine how best to manipulate data sources to get the answers you need.

EDA involves generating summary statistics for numerical data in the dataset and creating various graphical representations to understand the data better and make it more attractive and appealing.

# Data cleaning and visualization

### Data cleaning
Because our data set contains a high number of null values in the rating column, we remove them. Because some columns have fewer null values, we replace the null values in these columns with the mode value of that column. Our data set includes duplicate rows for a single application. We also remove duplicate rows because they contain the same data. Remove any rows with a rating greater than 5.
### Data Transforming
We can observe from the data frame information that all columns including rating have the object data type, however certain columns such as reviews, size, installs, and price have a number value. As a result, we must convert them to the right data type and remove any undesirable values from the numerical columns, such as '+' and ',' from installs and '$' from pricing. We have some values in KB and some values in MB in the size column, thus we transform all of the data in MB.

# Installs

```python
df['Installs'].head()
```

```python
#Removing all the commas and plus sign`
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',',''))
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+',''))
df['Installs'] = df['Installs'].astype(dtype = 'int')
```

```python
df['Installs'].unique()
```

```python
plt.figure(figsize=(12,8))
f = sns.countplot(df['Installs'], palette= "viridis" )
f.set_xticklabels(f.get_xticklabels(), rotation = 40)
plt.show()
```

We can see that the majority of apps fall into the 1M+ category.

```python
sorted_values = sorted(df['Installs'].unique())
df['Installs Classes'] = df['Installs'].replace(sorted_values, range(0,len(sorted_values)))
df['Installs Classes'].head()
```

```python
import scipy.stats as sp

plt.figure(figsize=(13,13))
plt.subplot(2,2,1)
f = sns.kdeplot(df[df['Installs Classes'] == 5]['Rating'], shade = True, color = 'purple')
plt.title("Ratings variation for apps above 5 installs")
f.set_xticks([1,2,3,4,5])

plt.subplot(2,2,2)
f = sns.kdeplot(df[df['Installs Classes'] == 6]['Rating'], shade = True, color = 'purple')
plt.title("Ratings variation for apps above 500 installs")
f.set_xticks([1,2,3,4,5])

plt.subplot(2,2,3)
f = sns.kdeplot(df[df['Installs Classes'] == 17]['Rating'], shade = True, color = 'purple')
plt.title("Ratings variation for apps above 500M installs")
f.set_xticks([1,2,3,4,5])

plt.subplot(2,2,4)
f = sns.kdeplot(df[df['Installs Classes'] == 18]['Rating'], shade = True, color = 'purple')
plt.title("Ratings variation for apps above 1B installs")
f.set_xticks([1,2,3,4,5])

plt.show()
print("Variation in Rating of installs above 100 installs ",sp.variation(df[df['Installs Classes'] == 5]['Rating']))
print("Variation in Rating of installs above 500 installs ",sp.variation(df[df['Installs Classes'] == 6]['Rating']))
print("Variation in Rating of installs above 500M installs ",sp.variation(df[df['Installs Classes'] == 17]['Rating']))
print("Variation in Rating of installs above 1B installs ",sp.variation(df[df['Installs Classes'] == 18]['Rating']))
```

```python
df['Size'].head()
```

```python
print(df['Size'].unique())
```

Uncanny observation: rather than the actual size, a value's label reads "Varies with device." So, it would be best to substitute some mean or median values.

Two issues exist with this column:

It has values with the label "Varies with device." Its size includes Ms and ks. Values are of type string. The KB and MB size scales would be thrown off if we omitted the Ms and Ks. Hence, we convert MBs to KBs. This is one approach to doing this.

```python
df['Size'] = df['Size'].apply(lambda x: x.replace('M', '*1000'))
df['Size'] = df['Size'].apply(lambda x: x.replace('k', ''))
```

In order to enable the usage of eval on the column, use -1 as the signal for the values "Varies with device."

```python
df['Size'].replace('Varies with device', '-1', inplace = True)
```

```python
df['Size'] = df['Size'].apply(lambda x: eval(x))
```

```python
df['Size'] = df['Size'].replace(-1,np.nan) #Changing the values to null then we can fill them with mean value
df['Size'].fillna(np.mean(df['Size']), inplace = True)
```

```python
plt.figure(figsize=(18,9))
sns.distplot(df['Size'])
plt.xlabel('Size in KBs')
plt.xticks(list(range(0, int(max(df['Size'])), 5000)))
plt.show
```

We can observe from the graph above that programmes that are 4-5MB or 20–25MB in size are more likely to be downloaded.

```python
plt.figure(figsize=(12,12))
sns.regplot(y=df['Size'],x=df['Installs Classes'],color='black')
plt.title('Size vs Installs')
plt.show()
```
